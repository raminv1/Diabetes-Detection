```python
#selecting target variables
X = df.drop('Outcome', axis=1)
y = df['Outcome']
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

###### we are trying to import KNN classifier from sklearn

```python
knn = KNeighborsClassifier(n_neighbors=20)
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
labels = sorted(list(set(y_test)))
cm = confusion_matrix(y_test, pred, labels=labels)
ax = sns.heatmap(cm, annot=True, fmt='d', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

print(classification_report(y_test,pred))

```

###### From above classification model we can see,accuracy of f1 score is around 79%, macro avg is 73% amd weighted avg is 78%.which is best when compared to other models.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense 
```

```python
model = Sequential()
```

```python
#Ref. for adding layers and number of neurons 
#31 features + 1 for bias 
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
#model.add(Dense(8,activation='relu'))

model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=600)
```

```python
loss = pd.DataFrame(model.history.history)
loss.head()
```

```python
loss.plot()
```

```python
model = Sequential() 

model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlystop = EarlyStopping(monitor='val_loss',mode='min',patience=20)
```

```python
model.fit(X_train,y_train,
          validation_data=(X_test,y_test),
          epochs=600, 
          callbacks=[earlystop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
#helps overfitting problem 
# you can set the dropout rate 
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential() 

model.add(Dense(32,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(16,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlystop = EarlyStopping(monitor='val_loss',mode='min',patience=20)
```

```python
model.fit(X_train,y_train,
          validation_data=(X_test,y_test),
          epochs=600, 
          callbacks=[earlystop])
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
predict = (model.predict(X_test) > 0.5).astype("int32")
```

```python
print(confusion_matrix(y_test,predict))
```

```python
print(classification_report(y_test,predict))
```

###### From above classification model we can see,accuracy of f1 score is around 77%, macro avg is 72% amd weighted avg is 76%.which is not best when compared to other models. Because other models have 79% accuracy.
